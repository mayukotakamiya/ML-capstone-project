```python
#Load packages required for loading the npy files.
import numpy as np
import pandas as pd

# #Load the npy file called "data.npy"
Initial_input_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_7/initial_inputs.npy")
Initial_output_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_7/initial_outputs.npy")

#I want to know dimentionality of the initial_input_data  
Initial_input_data.shape


```

```python

# X and y already loaded
X = Initial_input_data
y = Initial_output_data

def propose_next_point(X, y, n_candidates=8000, w_exploit=0.6, w_explore=0.4, seed=42):
    """
    Simple explore/exploit heuristic:
    - exploit: sample near best observed point
    - explore: sample far from existing samples
    """

    rng = np.random.default_rng(seed)
    n, d = X.shape

    # Best observed point so far
    best_x = X[np.argmax(y)]

    # Determine sampling bounds from existing data
    lower_bounds = X.min(axis=0)
    upper_bounds = X.max(axis=0)

    # Generate random candidate points within observed bounds
    C = rng.uniform(lower_bounds, upper_bounds, size=(n_candidates, d))

    # Exploitation term (closer to best is better)
    dist_to_best = np.linalg.norm(C - best_x, axis=1)
    exploit = 1 - (dist_to_best / (dist_to_best.max() + 1e-12))

    # Exploration term (farther from existing points is better)
    dists = np.sqrt(((C[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
    min_dist = dists.min(axis=1)
    explore = min_dist / (min_dist.max() + 1e-12)

    # Combined score
    score = w_exploit * exploit + w_explore * explore

    return C[np.argmax(score)]

def format_query(x):
    return "-".join([f"{v:0.6f}" for v in x])

# Generate next query
next_x = propose_next_point(X, y)

print("Submit this to the portal:")
print(format_query(next_x))
```

```python
#Re-run the loop
X=Combined_input_data
y=Combined_output_data

def propose_next_point(X, y, n_candidates=8000, w_exploit=0.75, w_explore=0.25, seed=234):
    """
    Simple explore/exploit heuristic:
    - exploit: sample near best observed point
    - explore: sample far from existing samples
    """

    rng = np.random.default_rng(seed)
    n, d = X.shape

    # Best observed point so far
    best_x = X[np.argmax(y)]
    
    radius = 0.05  # small local refinement

    # Determine sampling bounds from existing data
    lower = np.maximum(best_x - radius, X.min(axis=0))
    upper = np.minimum(best_x + radius, X.max(axis=0))


    # Generate random candidate points within observed bounds
    C = rng.uniform(lower, upper, size=(n_candidates, d))

    # Exploitation term (closer to best is better)
    dist_to_best = np.linalg.norm(C - best_x, axis=1)
    exploit = 1 - (dist_to_best / (dist_to_best.max() + 1e-12))

    # Exploration term (farther from existing points is better)
    dists = np.sqrt(((C[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
    min_dist = dists.min(axis=1)
    explore = min_dist / (min_dist.max() + 1e-12)

    # Combined score
    score = w_exploit * exploit + w_explore * explore

    return C[np.argmax(score)]

def format_query(x):
    return "-".join([f"{v:0.6f}" for v in x])

# Generate next query
next_x = propose_next_point(X, y)

print("Submit this to the portal:")
print(format_query(next_x))
```

```python
##WEEK 4- SVM and NN hybrid

import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neural_network import MLPRegressor

# --------------------------------------------------
# 1. Load dataset
# --------------------------------------------------

X = Combined_input_data4
y = Combined_output_data4

print("Dataset shape:", X.shape)

# --------------------------------------------------
# 2. Convert outputs to classification labels
# --------------------------------------------------

threshold = np.quantile(y, 0.7)   # top 30% = good
labels = (y >= threshold).astype(int)

# --------------------------------------------------
# 3. Scale inputs
# --------------------------------------------------

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --------------------------------------------------
# 4. Train SVM classifier
# --------------------------------------------------

svm = SVC(kernel="rbf", C=1.0, gamma="scale")
svm.fit(X_scaled, labels)

# --------------------------------------------------
# 5. Train neural network surrogate
# --------------------------------------------------

nn = MLPRegressor(
    hidden_layer_sizes=(24,12),
    activation="relu",
    solver="adam",
    max_iter=2000,
    random_state=42
)

nn.fit(X_scaled, y)

# --------------------------------------------------
# 6. Generate candidate points
# --------------------------------------------------

rng = np.random.default_rng(42)

lower = X.min(axis=0)
upper = X.max(axis=0)

candidates = rng.uniform(lower, upper, size=(10000, X.shape[1]))

# --------------------------------------------------
# 7. Remove candidates too close to existing points
# --------------------------------------------------

dist = np.sqrt(((candidates[:,None,:] - X[None,:,:])**2).sum(axis=2))
min_dist = dist.min(axis=1)

mask = min_dist > 0.01
candidates = candidates[mask]

# --------------------------------------------------
# 8. Score candidates using SVM + NN
# --------------------------------------------------

candidates_scaled = scaler.transform(candidates)

# SVM decision score (region quality)
svm_scores = svm.decision_function(candidates_scaled)

# NN predicted output
nn_predictions = nn.predict(candidates_scaled)

# Combine both scores
combined_score = nn_predictions + 0.1 * svm_scores

# --------------------------------------------------
# 9. Select best candidate
# --------------------------------------------------

best_candidate = candidates[np.argmax(combined_score)]

# --------------------------------------------------
# 10. Format for portal
# --------------------------------------------------

query_string = "-".join([f"{v:0.6f}" for v in best_candidate])

print("Next query:")
print(query_string)
```

```python
##WEEK 5 - train a small neural network surrogate
#$generate both global and local candidates
#use the NN’s predicted output to rank them
#add a small exploration bonus
#avoid querying points too close to existing samples

import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

# -------------------------------------------------
# 1. Load dataset
# -------------------------------------------------

X = Combined_input_data5
y = Combined_output_data5

print("Dataset shape:", X.shape)

# -------------------------------------------------
# 2. Scale inputs
# -------------------------------------------------

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------------------------
# 3. Train neural network surrogate
# -------------------------------------------------

model = MLPRegressor(
    hidden_layer_sizes=(40, 20),   # a bit larger for 6D
    activation="relu",
    solver="adam",
    alpha=1e-3,                    # light regularisation
    max_iter=4000,
    random_state=42
)

model.fit(X_scaled, y)

# -------------------------------------------------
# 4. Identify current best point
# -------------------------------------------------

best_idx = np.argmax(y)
best_x = X[best_idx]

print("Best value so far:", y[best_idx])
print("Best point so far:", best_x)

# -------------------------------------------------
# 5. Generate candidate points
# -------------------------------------------------

rng = np.random.default_rng(42)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Global exploration candidates
global_candidates = rng.uniform(lower, upper, size=(8000, 6))

# Local refinement candidates around current best
local_candidates = rng.normal(loc=best_x, scale=0.04, size=(8000, 6))
local_candidates = np.clip(local_candidates, lower, upper)

# Combine candidates
candidates = np.vstack([global_candidates, local_candidates])

# -------------------------------------------------
# 6. Remove near-duplicate candidates
# -------------------------------------------------

dist = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist = dist.min(axis=1)

mask = min_dist > 0.01
candidates = candidates[mask]
min_dist = min_dist[mask]

print("Candidates after filtering:", candidates.shape[0])

# -------------------------------------------------
# 7. Predict outputs with neural network
# -------------------------------------------------

candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)

# -------------------------------------------------
# 8. Add exploration bonus
# -------------------------------------------------

explore_score = min_dist / (min_dist.max() + 1e-12)

# Balanced score: prediction + exploration bonus
score = predictions + 0.10 * explore_score

# -------------------------------------------------
# 9. Select best candidate
# -------------------------------------------------

best_candidate_idx = np.argmax(score)
next_query = candidates[best_candidate_idx]

print("Predicted score of next query:", predictions[best_candidate_idx])

# -------------------------------------------------
# 10. Format for portal submission
# -------------------------------------------------

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
##WEEK 6 
#Use a slightly larger neural network surrogate
#Combine:
#global exploration (important in 6D)
#local refinement
#Add exploration bonus
#Avoid over-exploitation (previous stagnation risk)

import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

# -------------------------------------------------
# 1. Load dataset
# -------------------------------------------------

X = Combined_input_data6
y = Combined_output_data6

print("Dataset shape:", X.shape)

# -------------------------------------------------
# 2. Scale inputs
# -------------------------------------------------

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------------------------
# 3. Train neural network surrogate
# -------------------------------------------------

model = MLPRegressor(
    hidden_layer_sizes=(40, 20),   # slightly larger for 6D
    activation="relu",
    solver="adam",
    alpha=1e-3,                    # regularisation
    max_iter=4000,
    random_state=42
)

model.fit(X_scaled, y)

# -------------------------------------------------
# 4. Identify best point
# -------------------------------------------------

best_idx = np.argmax(y)
best_x = X[best_idx]

print("Best value:", y[best_idx])

# -------------------------------------------------
# 5. Generate candidate points
# -------------------------------------------------

rng = np.random.default_rng(123)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Global exploration (important in 6D)
global_candidates = rng.uniform(lower, upper, size=(8000, 6))

# Local refinement around best point
local_candidates = rng.normal(loc=best_x, scale=0.04, size=(8000, 6))
local_candidates = np.clip(local_candidates, lower, upper)

# Combine candidates
candidates = np.vstack([global_candidates, local_candidates])

# -------------------------------------------------
# 6. Remove near-duplicate candidates
# -------------------------------------------------

dist = np.sqrt(((candidates[:, None, :] - X[None, :, :])**2).sum(axis=2))
min_dist = dist.min(axis=1)

mask = min_dist > 0.01
candidates = candidates[mask]
min_dist = min_dist[mask]

print("Candidates after filtering:", candidates.shape[0])

# -------------------------------------------------
# 7. Predict outputs
# -------------------------------------------------

candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)

# -------------------------------------------------
# 8. Balanced scoring (avoid overfitting)
# -------------------------------------------------

explore_score = min_dist / (min_dist.max() + 1e-12)

# Balanced approach (slightly exploration-aware)
score = predictions + 0.12 * explore_score

# -------------------------------------------------
# 9. Select next query
# -------------------------------------------------

best_candidate_idx = np.argmax(score)
next_query = candidates[best_candidate_idx]

print("Predicted best value:", predictions[best_candidate_idx])

# -------------------------------------------------
# 10. Format for portal
# -------------------------------------------------

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```
