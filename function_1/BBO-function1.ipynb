```python
#Load packages required for loading the npy files.
import numpy as np
import pandas as pd

# #Load the npy file called "data.npy"
Initial_input_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_1/initial_inputs.npy")
Initial_output_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_1/initial_outputs.npy")

#I want to know dimentionality of the initial_input_data  
Initial_input_data.shape


```

```python

# X and y already loaded
X = Initial_input_data
y = Initial_output_data

def propose_next_point(X, y, n_candidates=8000, w_exploit=0.6, w_explore=0.4, seed=42):
    """
    Simple explore/exploit heuristic:
    - exploit: sample near best observed point
    - explore: sample far from existing samples
    """

    rng = np.random.default_rng(seed)
    n, d = X.shape

    # Best observed point so far
    best_x = X[np.argmax(y)]

    # Determine sampling bounds from existing data
    lower_bounds = X.min(axis=0)
    upper_bounds = X.max(axis=0)

    # Generate random candidate points within observed bounds
    C = rng.uniform(lower_bounds, upper_bounds, size=(n_candidates, d))

    # Exploitation term (closer to best is better)
    dist_to_best = np.linalg.norm(C - best_x, axis=1)
    exploit = 1 - (dist_to_best / (dist_to_best.max() + 1e-12))

    # Exploration term (farther from existing points is better)
    dists = np.sqrt(((C[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
    min_dist = dists.min(axis=1)
    explore = min_dist / (min_dist.max() + 1e-12)

    # Combined score
    score = w_exploit * exploit + w_explore * explore

    return C[np.argmax(score)]

def format_query(x):
    return "-".join([f"{v:0.6f}" for v in x])

# Generate next query
next_x = propose_next_point(X, y)

print("Submit this to the portal:")
print(format_query(next_x))
```

```python
#WEEK 8 

X = Combined_input_data8
y = Combined_output_data8

print("Dataset shape:", X.shape)

# "Low temperature" -> more focused search
EXPLOIT_WEIGHT = 0.72
EXPLORE_WEIGHT = 0.28

# tighter local search radius
LOCAL_RADIUS = 0.020

# candidate budget
N_LOCAL = 5000
N_GLOBAL = 1200

# avoid near-duplicate queries
MIN_SEP = 0.01

# top-k style selection
TOP_K = 25

SEED = 456

best_idx = np.argmax(y)
best_x = X[best_idx]
best_y = y[best_idx]

# Use top 20% as "attention set"
threshold = np.quantile(y, 0.80)
top_points = X[y >= threshold]

print("Best value so far:", best_y)
print("Best point so far:", best_x)
print("Top attention points:", len(top_points))

# fallback if very few points qualify
if len(top_points) == 0:
    top_points = X[best_idx:best_idx+1]

rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# local candidates around current best
local_best = rng.normal(loc=best_x, scale=LOCAL_RADIUS, size=(N_LOCAL // 2, 2))

# local candidates around top-performing region
local_top = []
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS * 1.2, size=(max(50, N_LOCAL // (2 * len(top_points))), 2))
    local_top.append(local)
local_top = np.vstack(local_top)

# global exploration candidates
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 2))

# combine and clip to observed bounds
candidates = np.vstack([local_best, local_top, global_candidates])
candidates = np.clip(candidates, lower, upper)

dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])

# Exploitation: closeness to best point
dist_to_best = np.linalg.norm(candidates - best_x, axis=1)
exploit_score = 1 - dist_to_best / (dist_to_best.max() + 1e-12)

# Exploration: distance from known points
explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# Combined score
score = EXPLOIT_WEIGHT * exploit_score + EXPLORE_WEIGHT * explore_score

top_k_idx = np.argsort(score)[-TOP_K:]
top_k_candidates = candidates[top_k_idx]
top_k_scores = score[top_k_idx]
top_k_explore = explore_score[top_k_idx]

# choose the candidate with the best blend of score + novelty
final_score = top_k_scores + 0.05 * top_k_explore
next_query = top_k_candidates[np.argmax(final_score)]

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 9 

X = Combined_input_data9
y = Combined_output_data9

print("Dataset shape:", X.shape)

# Strong exploitation: enough data should now reveal the best region
EXPLOIT_WEIGHT = 0.80
EXPLORE_WEIGHT = 0.20

# Tight local search radius
LOCAL_RADIUS = 0.015

# Candidate budget
N_LOCAL_BEST = 5000
N_LOCAL_TOP = 2000
N_GLOBAL = 800

# Focus on strongest region
TOP_QUANTILE = 0.85   # top 15%
TOP_K = 15

# Avoid duplicate queries
MIN_SEP = 0.01

SEED = 789

best_idx = np.argmax(y)
best_x = X[best_idx]
best_y = y[best_idx]

threshold = np.quantile(y, TOP_QUANTILE)
top_points = X[y >= threshold]

print("Best value so far:", best_y)
print("Best point so far:", best_x)
print("Top attention points:", len(top_points))

if len(top_points) == 0:
    top_points = X[best_idx:best_idx + 1]

rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Very focused local search around best point
local_best = rng.normal(loc=best_x, scale=LOCAL_RADIUS, size=(N_LOCAL_BEST, 2))

# Small local search around other strong points
local_top = []
n_per_top = max(50, N_LOCAL_TOP // len(top_points))
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS * 1.1, size=(n_per_top, 2))
    local_top.append(local)
local_top = np.vstack(local_top)

# Small global component
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 2))

# Combine and clip
candidates = np.vstack([local_best, local_top, global_candidates])
candidates = np.clip(candidates, lower, upper)

dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])

# Exploitation: closeness to best point
dist_to_best = np.linalg.norm(candidates - best_x, axis=1)
exploit_score = 1 - dist_to_best / (dist_to_best.max() + 1e-12)

# Exploration: novelty relative to previous points
explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# Combined score
score = EXPLOIT_WEIGHT * exploit_score + EXPLORE_WEIGHT * explore_score

top_k_idx = np.argsort(score)[-TOP_K:]
shortlist = candidates[top_k_idx]
shortlist_score = score[top_k_idx]
shortlist_explore = explore_score[top_k_idx]

# Small novelty bonus inside the shortlist
final_score = shortlist_score + 0.03 * shortlist_explore
next_query = shortlist[np.argmax(final_score)]


query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 11
from sklearn.cluster import KMeans

X = Combined_input_data11
y = Combined_output_data11

threshold = np.quantile(y, 0.80)
top_X = X[y >= threshold]

k = min(2, len(top_X))
kmeans = KMeans(n_clusters=k, random_state=11, n_init=10)
labels = kmeans.fit_predict(top_X)

# Choose cluster with highest mean output
best_cluster = None
best_mean = -np.inf
for c in range(k):
    cluster_points = top_X[labels == c]
    cluster_y = y[y >= threshold][labels == c]
    if cluster_y.mean() > best_mean:
        best_mean = cluster_y.mean()
        best_cluster = cluster_points

centroid = best_cluster.mean(axis=0)

# Move slightly outward from centroid toward farthest point in cluster
dists = np.linalg.norm(best_cluster - centroid, axis=1)
edge_point = best_cluster[np.argmax(dists)]
direction = edge_point - centroid
direction = direction / (np.linalg.norm(direction) + 1e-12)

next_query = centroid + 0.02 * direction
next_query = np.clip(next_query, X.min(axis=0), X.max(axis=0))

print("-".join([f"{v:0.6f}" for v in next_query]))
```
