```python
#Load packages required for loading the npy files.
import numpy as np
import pandas as pd

# #Load the npy file called "data.npy"
Initial_input_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_4/initial_inputs.npy")
Initial_output_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_4/initial_outputs.npy")

#I want to know dimentionality of the initial_input_data  
Initial_input_data.shape


```

```python

# X and y already loaded
X = Initial_input_data
y = Initial_output_data

def propose_next_point(X, y, n_candidates=8000, w_exploit=0.6, w_explore=0.4, seed=42):
    """
    Simple explore/exploit heuristic:
    - exploit: sample near best observed point
    - explore: sample far from existing samples
    """

    rng = np.random.default_rng(seed)
    n, d = X.shape

    # Best observed point so far
    best_x = X[np.argmax(y)]

    # Determine sampling bounds from existing data
    lower_bounds = X.min(axis=0)
    upper_bounds = X.max(axis=0)

    # Generate random candidate points within observed bounds
    C = rng.uniform(lower_bounds, upper_bounds, size=(n_candidates, d))

    # Exploitation term (closer to best is better)
    dist_to_best = np.linalg.norm(C - best_x, axis=1)
    exploit = 1 - (dist_to_best / (dist_to_best.max() + 1e-12))

    # Exploration term (farther from existing points is better)
    dists = np.sqrt(((C[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
    min_dist = dists.min(axis=1)
    explore = min_dist / (min_dist.max() + 1e-12)

    # Combined score
    score = w_exploit * exploit + w_explore * explore

    return C[np.argmax(score)]

def format_query(x):
    return "-".join([f"{v:0.6f}" for v in x])

# Generate next query
next_x = propose_next_point(X, y)

print("Submit this to the portal:")
print(format_query(next_x))
```

```python
#week 8

X = Combined_input_data8
y = Combined_output_data8

print("Dataset shape:", X.shape)

HIDDEN_LAYERS = (24, 12)
ALPHA = 1e-3
LOCAL_RADIUS = 0.032

# candidate budget
N_LOCAL_BEST = 3500
N_LOCAL_TOP = 3000
N_GLOBAL = 3500

# attention + shortlist settings
TOP_QUANTILE = 0.80   # top 20% attention set
TOP_K = 35
EXPLORE_BONUS = 0.10

# near-duplicate protection
MIN_SEP = 0.01

SEED = 456
MAX_ITER = 3500

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


model = MLPRegressor(
    hidden_layer_sizes=HIDDEN_LAYERS,
    activation="relu",
    solver="adam",
    alpha=ALPHA,
    max_iter=MAX_ITER,
    random_state=SEED
)

model.fit(X_scaled, y)

best_idx = np.argmax(y)
best_x = X[best_idx]
best_y = y[best_idx]

threshold = np.quantile(y, TOP_QUANTILE)
top_points = X[y >= threshold]

print("Best value so far:", best_y)
print("Best point so far:", best_x)
print("Top attention points:", len(top_points))

if len(top_points) == 0:
    top_points = X[best_idx:best_idx + 1]

rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Local around current best
local_best = rng.normal(loc=best_x, scale=LOCAL_RADIUS, size=(N_LOCAL_BEST, 4))

# Local around top-performing region
local_top = []
n_per_top = max(100, N_LOCAL_TOP // len(top_points))
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS * 1.1, size=(n_per_top, 4))
    local_top.append(local)
local_top = np.vstack(local_top)

# Global exploration
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 4))

# Combine and clip
candidates = np.vstack([local_best, local_top, global_candidates])
candidates = np.clip(candidates, lower, upper)

dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])

candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)

explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# Main score: surrogate output + exploration bonus
score = predictions + EXPLORE_BONUS * explore_score

# top-k shortlist
top_k_idx = np.argsort(score)[-TOP_K:]
shortlist = candidates[top_k_idx]
shortlist_score = score[top_k_idx]
shortlist_explore = explore_score[top_k_idx]

# final score slightly rewards diversity among strong candidates
final_score = shortlist_score + 0.05 * shortlist_explore
next_query = shortlist[np.argmax(final_score)]

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 9

X = Combined_input_data9
y = Combined_output_data9

print("Dataset shape:", X.shape)

# Moderate surrogate, now trusted more than before
HIDDEN_LAYERS = (24, 12)
ALPHA = 1e-3
LOCAL_RADIUS = 0.028

# Candidate budget
N_LOCAL_BEST = 4200
N_LOCAL_TOP = 2800
N_GLOBAL = 2200

# More exploitive than Query 8, but still not greedy
TOP_QUANTILE = 0.82   # top 18% attention set
TOP_K = 25
EXPLORE_BONUS = 0.08

# Duplicate protection
MIN_SEP = 0.01

SEED = 789
MAX_ITER = 4000

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


model = MLPRegressor(
    hidden_layer_sizes=HIDDEN_LAYERS,
    activation="relu",
    solver="adam",
    alpha=ALPHA,
    max_iter=MAX_ITER,
    random_state=SEED
)

model.fit(X_scaled, y)

best_idx = np.argmax(y)
best_x = X[best_idx]
best_y = y[best_idx]

threshold = np.quantile(y, TOP_QUANTILE)
top_points = X[y >= threshold]

print("Best value so far:", best_y)
print("Best point so far:", best_x)
print("Top attention points:", len(top_points))

if len(top_points) == 0:
    top_points = X[best_idx:best_idx + 1]


rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Local search around current best
local_best = rng.normal(loc=best_x, scale=LOCAL_RADIUS, size=(N_LOCAL_BEST, 4))

# Local search around top-performing region
local_top = []
n_per_top = max(80, N_LOCAL_TOP // len(top_points))
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS * 1.08, size=(n_per_top, 4))
    local_top.append(local)
local_top = np.vstack(local_top)

# Global exploration
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 4))

# Combine and clip
candidates = np.vstack([local_best, local_top, global_candidates])
candidates = np.clip(candidates, lower, upper)

dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])

candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)

explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# Surrogate-led score with modest novelty protection
score = predictions + EXPLORE_BONUS * explore_score

top_k_idx = np.argsort(score)[-TOP_K:]
shortlist = candidates[top_k_idx]
shortlist_score = score[top_k_idx]
shortlist_explore = explore_score[top_k_idx]

# Small novelty preference inside the shortlist
final_score = shortlist_score + 0.04 * shortlist_explore
next_query = shortlist[np.argmax(final_score)]


query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```
