```python
#Load packages required for loading the npy files.
import numpy as np
import pandas as pd

# #Load the npy file called "data.npy"
Initial_input_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_2/initial_inputs.npy")
Initial_output_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_2/initial_outputs.npy")

#I want to know dimentionality of the initial_input_data  
Initial_input_data.shape


```

```python

# X and y already loaded
X = Initial_input_data
y = Initial_output_data

def propose_next_point(X, y, n_candidates=8000, w_exploit=0.6, w_explore=0.4, seed=42):
    """
    Simple explore/exploit heuristic:
    - exploit: sample near best observed point
    - explore: sample far from existing samples
    """

    rng = np.random.default_rng(seed)
    n, d = X.shape

    # Best observed point so far
    best_x = X[np.argmax(y)]

    # Determine sampling bounds from existing data
    lower_bounds = X.min(axis=0)
    upper_bounds = X.max(axis=0)

    # Generate random candidate points within observed bounds
    C = rng.uniform(lower_bounds, upper_bounds, size=(n_candidates, d))

    # Exploitation term (closer to best is better)
    dist_to_best = np.linalg.norm(C - best_x, axis=1)
    exploit = 1 - (dist_to_best / (dist_to_best.max() + 1e-12))

    # Exploration term (farther from existing points is better)
    dists = np.sqrt(((C[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
    min_dist = dists.min(axis=1)
    explore = min_dist / (min_dist.max() + 1e-12)

    # Combined score
    score = w_exploit * exploit + w_explore * explore

    return C[np.argmax(score)]

def format_query(x):
    return "-".join([f"{v:0.6f}" for v in x])

# Generate next query
next_x = propose_next_point(X, y)

print("Submit this to the portal:")
print(format_query(next_x))
```

```python
#WEEK 8 

X = Combined_input_data8
y = Combined_output_data8

print("Dataset shape:", X.shape)

# "Higher temperature" -> more exploration because of noise
EXPLOIT_WEIGHT = 0.42
EXPLORE_WEIGHT = 0.58

# top region used as an "attention set"
TOP_QUANTILE = 0.70   # top 30%

# wider local radius than Function 1 because of noise
LOCAL_RADIUS = 0.040

# candidate budget
N_GLOBAL = 5000
N_LOCAL_PER_TOP = 600

# avoid near-duplicate queries
MIN_SEP = 0.01

# top-k shortlist size
TOP_K = 40

SEED = 456

threshold = np.quantile(y, TOP_QUANTILE)
top_points = X[y >= threshold]

print("Threshold for top region:", threshold)
print("Top attention points:", len(top_points))

if len(top_points) == 0:
    best_idx = np.argmax(y)
    top_points = X[best_idx:best_idx+1]

rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# global exploration candidates
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 2))

# local candidates around each top-performing point
local_candidates = []
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS, size=(N_LOCAL_PER_TOP, 2))
    local_candidates.append(local)

local_candidates = np.vstack(local_candidates)

# combine and clip to bounds
candidates = np.vstack([global_candidates, local_candidates])
candidates = np.clip(candidates, lower, upper)

dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])

# Exploitation: closeness to top-performing region
dist_to_top = np.sqrt(((candidates[:, None, :] - top_points[None, :, :]) ** 2).sum(axis=2))
min_dist_top = dist_to_top.min(axis=1)
exploit_score = 1 - min_dist_top / (min_dist_top.max() + 1e-12)

# Exploration: distance from known points
explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# Combined score
score = EXPLOIT_WEIGHT * exploit_score + EXPLORE_WEIGHT * explore_score

# top-k shortlist
top_k_idx = np.argsort(score)[-TOP_K:]
shortlist = candidates[top_k_idx]
shortlist_score = score[top_k_idx]
shortlist_explore = explore_score[top_k_idx]

# final selection prefers both quality and novelty
final_score = shortlist_score + 0.08 * shortlist_explore
next_query = shortlist[np.argmax(final_score)]

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 9 

X = Combined_input_data9
y = Combined_output_data9

print("Dataset shape:", X.shape)

# Moderate exploitation because noise is still important
EXPLOIT_WEIGHT = 0.55
EXPLORE_WEIGHT = 0.45

# Regional attention rather than single-point trust
TOP_QUANTILE = 0.75   # top 25% = good region

# Slightly wider local search due to noise
LOCAL_RADIUS = 0.030

# Candidate budget
N_LOCAL_TOP = 3500
N_LOCAL_BEST = 1500
N_GLOBAL = 2500

# Shortlist + duplicate protection
TOP_K = 25
MIN_SEP = 0.01

SEED = 789

best_idx = np.argmax(y)
best_x = X[best_idx]
best_y = y[best_idx]

threshold = np.quantile(y, TOP_QUANTILE)
top_points = X[y >= threshold]

print("Best value so far:", best_y)
print("Best point so far:", best_x)
print("Top-region points:", len(top_points))

if len(top_points) == 0:
    top_points = X[best_idx:best_idx + 1]

rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Local candidates around best point
local_best = rng.normal(loc=best_x, scale=LOCAL_RADIUS, size=(N_LOCAL_BEST, 2))

# Local candidates around robust top-performing region
local_top = []
n_per_top = max(80, N_LOCAL_TOP // len(top_points))
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS * 1.1, size=(n_per_top, 2))
    local_top.append(local)
local_top = np.vstack(local_top)

# Global exploration candidates
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 2))

# Combine and clip
candidates = np.vstack([local_best, local_top, global_candidates])
candidates = np.clip(candidates, lower, upper)


dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])


# Exploitation: closeness to the top-performing region
dist_to_top = np.sqrt(((candidates[:, None, :] - top_points[None, :, :]) ** 2).sum(axis=2))
min_dist_top = dist_to_top.min(axis=1)
exploit_score = 1 - min_dist_top / (min_dist_top.max() + 1e-12)

# Exploration: novelty relative to previous samples
explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# Combined score
score = EXPLOIT_WEIGHT * exploit_score + EXPLORE_WEIGHT * explore_score


top_k_idx = np.argsort(score)[-TOP_K:]
shortlist = candidates[top_k_idx]
shortlist_score = score[top_k_idx]
shortlist_explore = explore_score[top_k_idx]

# Slight novelty preference among strong candidates
final_score = shortlist_score + 0.05 * shortlist_explore
next_query = shortlist[np.argmax(final_score)]

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 11
from sklearn.cluster import KMeans

X = Combined_input_data11
y = Combined_output_data11

threshold = np.quantile(y, 0.70)
top_X = X[y >= threshold]
top_y = y[y >= threshold]

k = min(3, len(top_X))
kmeans = KMeans(n_clusters=k, random_state=12, n_init=10)
labels = kmeans.fit_predict(top_X)

best_cluster = None
best_score = -np.inf

for c in range(k):
    cluster_points = top_X[labels == c]
    cluster_y = top_y[labels == c]
    # prefer high mean and low variance
    score = cluster_y.mean() - 0.2 * cluster_y.std()
    if score > best_score:
        best_score = score
        best_cluster = cluster_points

centroid = best_cluster.mean(axis=0)

rng = np.random.default_rng(12)
candidates = rng.normal(loc=centroid, scale=0.03, size=(2000, 2))
candidates = np.clip(candidates, X.min(axis=0), X.max(axis=0))

dist = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
next_query = candidates[np.argmax(dist.min(axis=1))]

print("-".join([f"{v:0.6f}" for v in next_query]))
```
