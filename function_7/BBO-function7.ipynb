```python
#Load packages required for loading the npy files.
import numpy as np
import pandas as pd

# #Load the npy file called "data.npy"
Initial_input_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_7/initial_inputs.npy")
Initial_output_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_7/initial_outputs.npy")

#I want to know dimentionality of the initial_input_data  
Initial_input_data.shape


```

```python

# X and y already loaded
X = Initial_input_data
y = Initial_output_data

def propose_next_point(X, y, n_candidates=8000, w_exploit=0.6, w_explore=0.4, seed=42):
    """
    Simple explore/exploit heuristic:
    - exploit: sample near best observed point
    - explore: sample far from existing samples
    """

    rng = np.random.default_rng(seed)
    n, d = X.shape

    # Best observed point so far
    best_x = X[np.argmax(y)]

    # Determine sampling bounds from existing data
    lower_bounds = X.min(axis=0)
    upper_bounds = X.max(axis=0)

    # Generate random candidate points within observed bounds
    C = rng.uniform(lower_bounds, upper_bounds, size=(n_candidates, d))

    # Exploitation term (closer to best is better)
    dist_to_best = np.linalg.norm(C - best_x, axis=1)
    exploit = 1 - (dist_to_best / (dist_to_best.max() + 1e-12))

    # Exploration term (farther from existing points is better)
    dists = np.sqrt(((C[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
    min_dist = dists.min(axis=1)
    explore = min_dist / (min_dist.max() + 1e-12)

    # Combined score
    score = w_exploit * exploit + w_explore * explore

    return C[np.argmax(score)]

def format_query(x):
    return "-".join([f"{v:0.6f}" for v in x])

# Generate next query
next_x = propose_next_point(X, y)

print("Submit this to the portal:")
print(format_query(next_x))
```

```python
##WEEK 4- SVM and NN hybrid

from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neural_network import MLPRegressor

X = Combined_input_data4
y = Combined_output_data4

print("Dataset shape:", X.shape)

threshold = np.quantile(y, 0.7)   # top 30% = good
labels = (y >= threshold).astype(int)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

svm = SVC(kernel="rbf", C=1.0, gamma="scale")
svm.fit(X_scaled, labels)

nn = MLPRegressor(
    hidden_layer_sizes=(24,12),
    activation="relu",
    solver="adam",
    max_iter=2000,
    random_state=42
)

nn.fit(X_scaled, y)

rng = np.random.default_rng(42)

lower = X.min(axis=0)
upper = X.max(axis=0)

candidates = rng.uniform(lower, upper, size=(10000, X.shape[1]))

dist = np.sqrt(((candidates[:,None,:] - X[None,:,:])**2).sum(axis=2))
min_dist = dist.min(axis=1)

mask = min_dist > 0.01
candidates = candidates[mask]

candidates_scaled = scaler.transform(candidates)

# SVM decision score (region quality)
svm_scores = svm.decision_function(candidates_scaled)

# NN predicted output
nn_predictions = nn.predict(candidates_scaled)

# Combine both scores
combined_score = nn_predictions + 0.1 * svm_scores

best_candidate = candidates[np.argmax(combined_score)]

query_string = "-".join([f"{v:0.6f}" for v in best_candidate])

print("Next query:")
print(query_string)
```

```python
##WEEK 5 - train a small neural network surrogate
#$generate both global and local candidates
#use the NN’s predicted output to rank them
#add a small exploration bonus
#avoid querying points too close to existing samples


X = Combined_input_data5
y = Combined_output_data5

print("Dataset shape:", X.shape)


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = MLPRegressor(
    hidden_layer_sizes=(40, 20),   # a bit larger for 6D
    activation="relu",
    solver="adam",
    alpha=1e-3,                    # light regularisation
    max_iter=4000,
    random_state=42
)

model.fit(X_scaled, y)

best_idx = np.argmax(y)
best_x = X[best_idx]

print("Best value so far:", y[best_idx])
print("Best point so far:", best_x)

rng = np.random.default_rng(42)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Global exploration candidates
global_candidates = rng.uniform(lower, upper, size=(8000, 6))

# Local refinement candidates around current best
local_candidates = rng.normal(loc=best_x, scale=0.04, size=(8000, 6))
local_candidates = np.clip(local_candidates, lower, upper)

# Combine candidates
candidates = np.vstack([global_candidates, local_candidates])

dist = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist = dist.min(axis=1)

mask = min_dist > 0.01
candidates = candidates[mask]
min_dist = min_dist[mask]

print("Candidates after filtering:", candidates.shape[0])

candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)

explore_score = min_dist / (min_dist.max() + 1e-12)

# Balanced score: prediction + exploration bonus
score = predictions + 0.10 * explore_score

best_candidate_idx = np.argmax(score)
next_query = candidates[best_candidate_idx]

print("Predicted score of next query:", predictions[best_candidate_idx])

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 8

X = Combined_input_data8
y = Combined_output_data8

print("Dataset shape:", X.shape)


# diversity-aware / slightly higher-temperature search
HIDDEN_LAYERS = (40, 20)
ALPHA = 1e-3
LOCAL_RADIUS = 0.040

# candidate budget
N_LOCAL_BEST = 3500
N_LOCAL_TOP = 4000
N_GLOBAL = 6500

# attention + shortlist settings
TOP_QUANTILE = 0.80   # top 20% attention set
TOP_K = 40
EXPLORE_BONUS = 0.12

# near-duplicate protection
MIN_SEP = 0.01

SEED = 456
MAX_ITER = 4500


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = MLPRegressor(
    hidden_layer_sizes=HIDDEN_LAYERS,
    activation="relu",
    solver="adam",
    alpha=ALPHA,
    max_iter=MAX_ITER,
    random_state=SEED
)

model.fit(X_scaled, y)

best_idx = np.argmax(y)
best_x = X[best_idx]
best_y = y[best_idx]

threshold = np.quantile(y, TOP_QUANTILE)
top_points = X[y >= threshold]

print("Best value so far:", best_y)
print("Best point so far:", best_x)
print("Top attention points:", len(top_points))

if len(top_points) == 0:
    top_points = X[best_idx:best_idx + 1]


rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# local candidates around current best
local_best = rng.normal(loc=best_x, scale=LOCAL_RADIUS, size=(N_LOCAL_BEST, 6))

# local candidates around top-performing region
local_top = []
n_per_top = max(100, N_LOCAL_TOP // len(top_points))
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS * 1.1, size=(n_per_top, 6))
    local_top.append(local)
local_top = np.vstack(local_top)

# global exploration candidates
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 6))

# combine and clip
candidates = np.vstack([local_best, local_top, global_candidates])
candidates = np.clip(candidates, lower, upper)


dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])


candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)


explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# score = model prediction + novelty bonus
score = predictions + EXPLORE_BONUS * explore_score


top_k_idx = np.argsort(score)[-TOP_K:]
shortlist = candidates[top_k_idx]
shortlist_score = score[top_k_idx]
shortlist_explore = explore_score[top_k_idx]

# slightly favour novelty inside the shortlist
final_score = shortlist_score + 0.06 * shortlist_explore
next_query = shortlist[np.argmax(final_score)]


query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 9 

X = Combined_input_data9
y = Combined_output_data9

print("Dataset shape:", X.shape)

# More trust in the surrogate than earlier rounds,
# but still regularised because 6D remains sparse
HIDDEN_LAYERS = (40, 20)
ALPHA = 1e-3
LOCAL_RADIUS = 0.035

# Candidate budget
N_LOCAL_BEST = 4200
N_LOCAL_TOP = 3200
N_GLOBAL = 3800

# Attention to strongest region, but not too narrow
TOP_QUANTILE = 0.82   # top 18%
TOP_K = 30
EXPLORE_BONUS = 0.10

# Duplicate protection
MIN_SEP = 0.01

SEED = 789
MAX_ITER = 5000


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = MLPRegressor(
    hidden_layer_sizes=HIDDEN_LAYERS,
    activation="relu",
    solver="adam",
    alpha=ALPHA,
    max_iter=MAX_ITER,
    random_state=SEED
)

model.fit(X_scaled, y)

best_idx = np.argmax(y)
best_x = X[best_idx]
best_y = y[best_idx]

threshold = np.quantile(y, TOP_QUANTILE)
top_points = X[y >= threshold]

print("Best value so far:", best_y)
print("Best point so far:", best_x)
print("Top attention points:", len(top_points))

if len(top_points) == 0:
    top_points = X[best_idx:best_idx + 1]

rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Local refinement around current best
local_best = rng.normal(loc=best_x, scale=LOCAL_RADIUS, size=(N_LOCAL_BEST, 6))

# Local refinement around other strong points
local_top = []
n_per_top = max(90, N_LOCAL_TOP // len(top_points))
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS * 1.08, size=(n_per_top, 6))
    local_top.append(local)
local_top = np.vstack(local_top)

# Global exploration
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 6))

# Combine and clip
candidates = np.vstack([local_best, local_top, global_candidates])
candidates = np.clip(candidates, lower, upper)

dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])

candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)

explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# Surrogate-led score with moderate diversity protection
score = predictions + EXPLORE_BONUS * explore_score

top_k_idx = np.argsort(score)[-TOP_K:]
shortlist = candidates[top_k_idx]
shortlist_score = score[top_k_idx]
shortlist_explore = explore_score[top_k_idx]

# Slight novelty preference inside the shortlist
final_score = shortlist_score + 0.05 * shortlist_explore
next_query = shortlist[np.argmax(final_score)]

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 11
from sklearn.cluster import KMeans

X = Combined_input_data11
y = Combined_output_data11

threshold = np.quantile(y, 0.75)
top_X = X[y >= threshold]
top_y = y[y >= threshold]

k = min(4, len(top_X))
kmeans = KMeans(n_clusters=k, random_state=17, n_init=10)
labels = kmeans.fit_predict(top_X)

cluster_info = []
for c in range(k):
    cluster_points = top_X[labels == c]
    cluster_y = top_y[labels == c]
    cluster_info.append((cluster_y.mean(), cluster_points.mean(axis=0), cluster_points))

cluster_info.sort(key=lambda t: t[0], reverse=True)
best_centroid = cluster_info[0][1]

rng = np.random.default_rng(17)
local = rng.normal(loc=best_centroid, scale=0.04, size=(3000, X.shape[1]))
global_ = rng.uniform(X.min(axis=0), X.max(axis=0), size=(3000, X.shape[1]))
candidates = np.vstack([local, global_])
candidates = np.clip(candidates, X.min(axis=0), X.max(axis=0))

dist = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist = dist.min(axis=1)

# choose candidate near best cluster but not too close to old points
dist_to_centroid = np.linalg.norm(candidates - best_centroid, axis=1)
score = -dist_to_centroid + 0.2 * min_dist
next_query = candidates[np.argmax(score)]

print("-".join([f"{v:0.6f}" for v in next_query]))
```
