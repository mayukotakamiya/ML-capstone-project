```python
#Load packages required for loading the npy files.
import numpy as np
import pandas as pd

# #Load the npy file called "data.npy"
Initial_input_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_8/initial_inputs.npy")
Initial_output_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_8/initial_outputs.npy")

#I want to know dimentionality of the initial_input_data  
Initial_input_data.shape


```

```python

# X and y already loaded
X = Initial_input_data
y = Initial_output_data

def propose_next_point(X, y, n_candidates=8000, w_exploit=0.6, w_explore=0.4, seed=42):
    """
    Simple explore/exploit heuristic:
    - exploit: sample near best observed point
    - explore: sample far from existing samples
    """

    rng = np.random.default_rng(seed)
    n, d = X.shape

    # Best observed point so far
    best_x = X[np.argmax(y)]

    # Determine sampling bounds from existing data
    lower_bounds = X.min(axis=0)
    upper_bounds = X.max(axis=0)

    # Generate random candidate points within observed bounds
    C = rng.uniform(lower_bounds, upper_bounds, size=(n_candidates, d))

    # Exploitation term (closer to best is better)
    dist_to_best = np.linalg.norm(C - best_x, axis=1)
    exploit = 1 - (dist_to_best / (dist_to_best.max() + 1e-12))

    # Exploration term (farther from existing points is better)
    dists = np.sqrt(((C[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
    min_dist = dists.min(axis=1)
    explore = min_dist / (min_dist.max() + 1e-12)

    # Combined score
    score = w_exploit * exploit + w_explore * explore

    return C[np.argmax(score)]

def format_query(x):
    return "-".join([f"{v:0.6f}" for v in x])

# Generate next query
next_x = propose_next_point(X, y)

print("Submit this to the portal:")
print(format_query(next_x))
```

```python
#WEEK 8

X = Combined_input_data8
y = Combined_output_data8

print("Dataset shape:", X.shape)

# higher-capacity but regularised surrogate
HIDDEN_LAYERS = (64, 32)
ALPHA = 1e-3
LOCAL_RADIUS = 0.045

# candidate budget
N_LOCAL_BEST = 4000
N_LOCAL_TOP = 5000
N_GLOBAL = 9000

# attention + shortlist settings
TOP_QUANTILE = 0.80   # top 20% attention set
TOP_K = 50
EXPLORE_BONUS = 0.15

# near-duplicate protection
MIN_SEP = 0.01

SEED = 456
MAX_ITER = 5000


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


model = MLPRegressor(
    hidden_layer_sizes=HIDDEN_LAYERS,
    activation="relu",
    solver="adam",
    alpha=ALPHA,
    max_iter=MAX_ITER,
    random_state=SEED
)

model.fit(X_scaled, y)

best_idx = np.argmax(y)
best_x = X[best_idx]
best_y = y[best_idx]

threshold = np.quantile(y, TOP_QUANTILE)
top_points = X[y >= threshold]

print("Best value so far:", best_y)
print("Best point so far:", best_x)
print("Top attention points:", len(top_points))

if len(top_points) == 0:
    top_points = X[best_idx:best_idx + 1]

rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# local candidates around current best
local_best = rng.normal(loc=best_x, scale=LOCAL_RADIUS, size=(N_LOCAL_BEST, 8))

# local candidates around top-performing region
local_top = []
n_per_top = max(120, N_LOCAL_TOP // len(top_points))
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS * 1.1, size=(n_per_top, 8))
    local_top.append(local)
local_top = np.vstack(local_top)

# global exploration candidates
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 8))

# combine and clip
candidates = np.vstack([local_best, local_top, global_candidates])
candidates = np.clip(candidates, lower, upper)

dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])

candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)

explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# score = surrogate prediction + exploration bonus
score = predictions + EXPLORE_BONUS * explore_score

top_k_idx = np.argsort(score)[-TOP_K:]
shortlist = candidates[top_k_idx]
shortlist_score = score[top_k_idx]
shortlist_explore = explore_score[top_k_idx]

# slightly favour novelty within the shortlist
final_score = shortlist_score + 0.07 * shortlist_explore
next_query = shortlist[np.argmax(final_score)]

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 9 

X = Combined_input_data9
y = Combined_output_data9

print("Dataset shape:", X.shape)

# Higher-capacity surrogate, but still regularised
HIDDEN_LAYERS = (64, 32)
ALPHA = 1e-3
LOCAL_RADIUS = 0.040

# Candidate budget
N_LOCAL_BEST = 4200
N_LOCAL_TOP = 4200
N_GLOBAL = 6500

# Focus on strongest region, but not too narrowly
TOP_QUANTILE = 0.82   # top 18%
TOP_K = 35
EXPLORE_BONUS = 0.12

# Duplicate protection
MIN_SEP = 0.01

SEED = 789
MAX_ITER = 5500

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

model = MLPRegressor(
    hidden_layer_sizes=HIDDEN_LAYERS,
    activation="relu",
    solver="adam",
    alpha=ALPHA,
    max_iter=MAX_ITER,
    random_state=SEED
)

model.fit(X_scaled, y)

best_idx = np.argmax(y)
best_x = X[best_idx]
best_y = y[best_idx]

threshold = np.quantile(y, TOP_QUANTILE)
top_points = X[y >= threshold]

print("Best value so far:", best_y)
print("Best point so far:", best_x)
print("Top attention points:", len(top_points))

if len(top_points) == 0:
    top_points = X[best_idx:best_idx + 1]

rng = np.random.default_rng(SEED)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Local refinement around current best
local_best = rng.normal(loc=best_x, scale=LOCAL_RADIUS, size=(N_LOCAL_BEST, 8))

# Local refinement around other strong points
local_top = []
n_per_top = max(100, N_LOCAL_TOP // len(top_points))
for pt in top_points:
    local = rng.normal(loc=pt, scale=LOCAL_RADIUS * 1.08, size=(n_per_top, 8))
    local_top.append(local)
local_top = np.vstack(local_top)

# Global exploration
global_candidates = rng.uniform(lower, upper, size=(N_GLOBAL, 8))

# Combine and clip
candidates = np.vstack([local_best, local_top, global_candidates])
candidates = np.clip(candidates, lower, upper)

dist_to_existing = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_existing = dist_to_existing.min(axis=1)

mask = min_dist_existing > MIN_SEP
candidates = candidates[mask]
min_dist_existing = min_dist_existing[mask]

print("Candidates after filtering:", candidates.shape[0])

candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)

explore_score = min_dist_existing / (min_dist_existing.max() + 1e-12)

# Surrogate-led score with meaningful diversity protection
score = predictions + EXPLORE_BONUS * explore_score

top_k_idx = np.argsort(score)[-TOP_K:]
shortlist = candidates[top_k_idx]
shortlist_score = score[top_k_idx]
shortlist_explore = explore_score[top_k_idx]

# Slight novelty preference within shortlist
final_score = shortlist_score + 0.06 * shortlist_explore
next_query = shortlist[np.argmax(final_score)]

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```

```python
#WEEK 11
from sklearn.cluster import KMeans

X = Combined_input_data11
y = Combined_output_data11

threshold = np.quantile(y, 0.80)
top_X = X[y >= threshold]
top_y = y[y >= threshold]

k = min(3, len(top_X))
kmeans = KMeans(n_clusters=k, random_state=18, n_init=10)
labels = kmeans.fit_predict(top_X)

centroids = []
scores = []

for c in range(k):
    cluster_points = top_X[labels == c]
    cluster_y = top_y[labels == c]
    centroids.append(cluster_points.mean(axis=0))
    scores.append(cluster_y.mean())

centroids = np.array(centroids)
scores = np.array(scores)
best_centroid = centroids[np.argmax(scores)]

rng = np.random.default_rng(18)

# mostly around best cluster, some global exploration
local = rng.normal(loc=best_centroid, scale=0.05, size=(4000, X.shape[1]))
global_ = rng.uniform(X.min(axis=0), X.max(axis=0), size=(5000, X.shape[1]))
candidates = np.vstack([local, global_])
candidates = np.clip(candidates, X.min(axis=0), X.max(axis=0))

dist_old = np.sqrt(((candidates[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
min_dist_old = dist_old.min(axis=1)

dist_centroid = np.linalg.norm(candidates - best_centroid, axis=1)

# balanced: close to best cluster, but still novel
score = -dist_centroid + 0.25 * min_dist_old
next_query = candidates[np.argmax(score)]

print("-".join([f"{v:0.6f}" for v in next_query]))
```
