```python
#Load packages required for loading the npy files.
import numpy as np
import pandas as pd

# #Load the npy file called "data.npy"
Initial_input_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_5/initial_inputs.npy")
Initial_output_data = np.load("/Users/mayukotakamiya/MachineLearning/initial_data/function_5/initial_outputs.npy")

#I want to know dimentionality of the initial_input_data  
Initial_input_data.shape


```

```python

# X and y already loaded
X = Initial_input_data
y = Initial_output_data

def propose_next_point(X, y, n_candidates=8000, w_exploit=0.6, w_explore=0.4, seed=42):
    """
    Simple explore/exploit heuristic:
    - exploit: sample near best observed point
    - explore: sample far from existing samples
    """

    rng = np.random.default_rng(seed)
    n, d = X.shape

    # Best observed point so far
    best_x = X[np.argmax(y)]

    # Determine sampling bounds from existing data
    lower_bounds = X.min(axis=0)
    upper_bounds = X.max(axis=0)

    # Generate random candidate points within observed bounds
    C = rng.uniform(lower_bounds, upper_bounds, size=(n_candidates, d))

    # Exploitation term (closer to best is better)
    dist_to_best = np.linalg.norm(C - best_x, axis=1)
    exploit = 1 - (dist_to_best / (dist_to_best.max() + 1e-12))

    # Exploration term (farther from existing points is better)
    dists = np.sqrt(((C[:, None, :] - X[None, :, :]) ** 2).sum(axis=2))
    min_dist = dists.min(axis=1)
    explore = min_dist / (min_dist.max() + 1e-12)

    # Combined score
    score = w_exploit * exploit + w_explore * explore

    return C[np.argmax(score)]

def format_query(x):
    return "-".join([f"{v:0.6f}" for v in x])

# Generate next query
next_x = propose_next_point(X, y)

print("Submit this to the portal:")
print(format_query(next_x))
```

```python
##WEEK 4 - NN + local optimisation

import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

# -------------------------------------------------
# 1. Load dataset (after previous queries)
# -------------------------------------------------

X = Combined_input_data4
y = Combined_output_data4

print("Dataset shape:", X.shape)

# -------------------------------------------------
# 2. Scale inputs
# -------------------------------------------------

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------------------------
# 3. Train neural network surrogate
# -------------------------------------------------

model = MLPRegressor(
    hidden_layer_sizes=(16,8),
    activation="relu",
    solver="adam",
    max_iter=2000,
    random_state=42
)

model.fit(X_scaled, y)

# -------------------------------------------------
# 4. Identify current best point
# -------------------------------------------------

best_idx = np.argmax(y)
best_x = X[best_idx]

print("Current best point:", best_x)

# -------------------------------------------------
# 5. Generate candidate points
# -------------------------------------------------

rng = np.random.default_rng(42)

# local candidates around best point
local_candidates = rng.normal(best_x, 0.05, size=(4000, X.shape[1]))

# global candidates (exploration)
lower = X.min(axis=0)
upper = X.max(axis=0)

global_candidates = rng.uniform(lower, upper, size=(4000, X.shape[1]))

candidates = np.vstack([local_candidates, global_candidates])

# clip values to valid range
candidates = np.clip(candidates, lower, upper)

# -------------------------------------------------
# 6. Avoid querying points too close to existing ones
# -------------------------------------------------

dist = np.sqrt(((candidates[:,None,:] - X[None,:,:])**2).sum(axis=2))
min_dist = dist.min(axis=1)

mask = min_dist > 0.01
candidates = candidates[mask]

# -------------------------------------------------
# 7. Predict outputs using the neural network
# -------------------------------------------------

candidates_scaled = scaler.transform(candidates)

predictions = model.predict(candidates_scaled)

# -------------------------------------------------
# 8. Choose best predicted candidate
# -------------------------------------------------

best_candidate = candidates[np.argmax(predictions)]

# -------------------------------------------------
# 9. Format query for portal
# -------------------------------------------------

query_string = "-".join([f"{v:0.6f}" for v in best_candidate])

print("Next query:")
print(query_string)
```

```python
#WEEK 5- Use a neural network (moderate depth), Focus more on local refinement, Still include global candidates, Slightly exploit-heavy scoring

import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor

# -------------------------------------------------
# 1. Load dataset
# -------------------------------------------------

X = Combined_input_data5
y = Combined_output_data5

print("Dataset shape:", X.shape)

# -------------------------------------------------
# 2. Scale inputs
# -------------------------------------------------

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# -------------------------------------------------
# 3. Train neural network surrogate
# -------------------------------------------------

model = MLPRegressor(
    hidden_layer_sizes=(24, 12),
    activation="relu",
    solver="adam",
    max_iter=2500,
    random_state=42
)

model.fit(X_scaled, y)

# -------------------------------------------------
# 4. Identify best point
# -------------------------------------------------

best_idx = np.argmax(y)
best_x = X[best_idx]

print("Best value:", y[best_idx])
print("Best point:", best_x)

# -------------------------------------------------
# 5. Generate candidate points
# -------------------------------------------------

rng = np.random.default_rng(42)

lower = X.min(axis=0)
upper = X.max(axis=0)

# Strong local refinement (main focus)
local_candidates = rng.normal(loc=best_x, scale=0.03, size=(6000, 4))
local_candidates = np.clip(local_candidates, lower, upper)

# Small global exploration
global_candidates = rng.uniform(lower, upper, size=(2000, 4))

# Combine
candidates = np.vstack([local_candidates, global_candidates])

# -------------------------------------------------
# 6. Remove near-duplicate candidates
# -------------------------------------------------

dist = np.sqrt(((candidates[:, None, :] - X[None, :, :])**2).sum(axis=2))
min_dist = dist.min(axis=1)

mask = min_dist > 0.01
candidates = candidates[mask]
min_dist = min_dist[mask]

# -------------------------------------------------
# 7. Predict outputs
# -------------------------------------------------

candidates_scaled = scaler.transform(candidates)
predictions = model.predict(candidates_scaled)

# -------------------------------------------------
# 8. Scoring (exploit-heavy)
# -------------------------------------------------

explore_score = min_dist / (min_dist.max() + 1e-12)

# Strong exploitation because unimodal
score = predictions + 0.05 * explore_score

# -------------------------------------------------
# 9. Select next query
# -------------------------------------------------

next_query = candidates[np.argmax(score)]

# -------------------------------------------------
# 10. Format for portal
# -------------------------------------------------

query_string = "-".join([f"{v:0.6f}" for v in next_query])

print("Next query:")
print(query_string)
```
